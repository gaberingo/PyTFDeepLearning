```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from keras.preprocessing import image

import matplotlib.pyplot as plt
```

```python
train_dir = '/home/gabe/Documents/TF_Learn_DL/data/10_food_classes_all_data/train'
test_dir = '/home/gabe/Documents/TF_Learn_DL/data/10_food_classes_all_data/test'

img = image.load_img(
    '/home/gabe/Documents/TF_Learn_DL/data/10_food_classes_all_data/train/pizza/1008104.jpg',
    target_size=(240,240),
    color_mode='rgb'
)
img_array = image.img_to_array(img)
img_array = tf.expand_dims(img_array, axis=0)/225.
```

```python
model = Sequential()
model.add(Conv2D(3, (3,3),activation='relu', input_shape = (240,240,3), padding='same'))
model.summary()
```

```python
new_image = model.predict(img_array)
```

```python
# Plotting the images
def plot_images(new_image):
    original_img = image.load_img(
    '/home/gabe/Documents/TF_Learn_DL/data/10_food_classes_all_data/train/pizza/1008104.jpg',
    target_size=(64,64)
)
    fig, axes = plt.subplots(1,2, figsize=(10,10))
    axes = axes.flatten()
    for i, img in enumerate([original_img, new_image]):
        axes[i].imshow(img)
        axes[i].set_xticks([10])
        axes[i].set_yticks([10])
    plt.tight_layout()
    plt.show()
    
```

```python
plot_images(new_image[0])
```
